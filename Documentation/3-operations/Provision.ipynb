```
var (user, profile) = await jinagaClient.Login();

var administrator = await jinagaClient.SingleUse(async principal =>
{
  var environment = await jinagaClient.Fact(new Environment(principal, "Development"));
  var administrator = await jinagaClient.Fact(new Administrator(user, environment, DateTime.UtcNow));
  return administrator;
});

jinagaClient.RenderFacts(administrator)
```
