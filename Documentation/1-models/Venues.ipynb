## Venue

The user can create, delete, and restore venues.

```
[FactType("Festify.Venue")]
public record Venue(Environment environment, Guid venueGuid)
{
  public Condition IsDeleted => Condition.Define(facts =>
    facts.Any<VenueDelete>(delete => delete.venue == this)
  );

  public Relation<VenueName> Names => Relation.Define(facts =>
    facts.OfType<VenueName>(name => name.venue == this &&
      !facts.Any<VenueName>(next => next.prior.Contains(name))
    )
  );
}

[FactType("Festify.Venue.Delete")]
public record VenueDelete(Venue venue);

[FactType("Festify.Venue.Restore")]
public record VenueRestore(VenueDelete delete);

[FactType("Festify.Venue.Name")]
public record VenueName(Venue venue, string value, VenueName[] prior);

Renderer.RenderTypes(typeof(VenueRestore), typeof(VenueName))
```
