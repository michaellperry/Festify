## Shows

A show occurs at a venue with an act at a particular date and time.

```
[FactType("Festify.Show")]
public record Show(Act act, Venue venue, DateTime start);

[FactType("Festify.Show.Delete")]
public record DeleteShow(Show show, DateTime deletedAt);

[FactType("Festify.Show.Restore")]
public record RestoreShow(DeleteShow delete);

Renderer.RenderTypes(typeof(Show), typeof(DeleteShow), typeof(RestoreShow))
```
