## Act

An Act represents an artist, group, or company that is the primary performer in an event.
The promoter can create, delete, and restore an act.
They can also update the act's title and image.

```
[FactType("Festify.Act")]
public record Act(Environment environment, Guid actGuid);

[FactType("Festify.Act.Title")]
public record ActTitle(Act act, string value, ActTitle[] prior);

[FactType("Festify.Act.Image")]
public record ActImage(Act act, string imageHash, ActImage[] prior);

[FactType("Festify.Act.Delete")]
public record ActDelete(Act act, DateTime deletedAt);

[FactType("Festify.Act.Restore")]
public record ActRestore(ActDelete delete);

Renderer.RenderTypes(typeof(Environment), typeof(Act), typeof(ActTitle), typeof(ActImage), typeof(ActDelete), typeof(ActRestore))
```
