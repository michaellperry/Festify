```
var environment = await jinagaClient.Fact(new Environment("test"));
```

```
var aaCenter = await jinagaClient.Fact(new Venue(environment, Guid.NewGuid()));
var aaName = await jinagaClient.Fact(new VenueName(aaCenter, "American Airlines Center", []));

jinagaClient.RenderFacts(aaName)
```

```
var broncoBowl = await jinagaClient.Fact(new Venue(environment, Guid.NewGuid()));
var broncoBowlName = await jinagaClient.Fact(new VenueName(broncoBowl, "Bronco Bowl", []));

jinagaClient.RenderFacts(broncoBowlName)
```

```
var broncoBowlDelete = await jinagaClient.Fact(new VenueDelete(broncoBowl));

jinagaClient.RenderFacts(broncoBowlName, broncoBowlDelete)
```

```
var venuesInEnvironment = Given<Environment>.Match((environment, facts) =>
  from venue in facts.OfType<Venue>()
  where venue.environment == environment && !venue.IsDeleted
  select new
  {
    Venue = venue,
    Names = venue.Names.Select(name => name.value)
  }
);

var venues = await jinagaClient.Query(venuesInEnvironment, environment);

venues
```

```
var starplex = await jinagaClient.Fact(new Venue(environment, Guid.NewGuid()));
```

```
var namesInVenue = Given<Venue>.Match((venue, facts) =>
  from name in facts.OfType<VenueName>()
  where name.venue == venue &&
    !facts.Any<VenueName>(next => next.prior.Contains(name))
  select name
);

var starplexNames = await jinagaClient.Query(namesInVenue, starplex);

jinagaClient.RenderFacts(starplexNames)
```

```
var desiredName = "Dos Equis Pavilion";

if (starplexNames.Count != 1 || starplexNames.Single().value != desiredName)
{
  await jinagaClient.Fact(new VenueName(starplex, desiredName, starplexNames.ToArray()));
}
```
